```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("/content/General_Stores.csv")
print("general Data:")
data.head()
```

```python
X=data[["Footfalls"]]
Y=data["DailySales"]
```

```python
plt.xlabel("Footfalls")
plt.ylabel("DailySales")
plt.scatter(X,Y,color="red",marker='o')
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
print(f"Coefficient (slope):{model.coef_}")
print(f"Intercept:{model.intercept_}")
```

```python
model.intercept_+ model.coef_*20
```

```python
model.predict([[35]])
```

```python
new_footfalls=pd.DataFrame({"Footfalls":[10,20,30,40,50]})
new_footfalls
```

```python
predicted_sales=model.predict(new_footfalls)
predicted_sales
```
