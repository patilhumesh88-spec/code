```python
# 📦 CELL 1: Setup & Imports (Run first)
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries loaded. Starting data pipeline...")
```

```python
# 🌦️ CELL 2: Fetch Hourly Weather from Open-Meteo (Free, No API Key)
BASE_URL = "https://archive-api.open-meteo.com/v1/archive"
PARAMS = {
    "latitude": 23.0225,      # Ahmedabad
    "longitude": 72.5714,
    "hourly": "temperature_2m,relative_humidity_2m,wind_speed_10m,cloud_cover",
    "start_date": "2022-01-01",
    "end_date": "2024-12-31",
    "timezone": "UTC"
}

print("⬇️ Downloading weather data...")
res = requests.get(BASE_URL, params=PARAMS)
res.raise_for_status()
weather = pd.DataFrame(res.json()['hourly'])
weather['timestamp'] = pd.to_datetime(weather['time'])
print(f"✅ Weather fetched: {len(weather)} hourly records")
```

```python
# ⚡ CELL 3: Build Demand Column (Realistic Proxy → Swap with Real Data Later)
# NOTE: Real Indian hourly demand is fragmented across state utilities/CEA.
# This creates a physically realistic baseline you can immediately replace.

print("⚙️ Generating demand pattern...")
weather['hour'] = weather['timestamp'].dt.hour

# Typical Indian grid curve: base + day/night + weekend dip
base_load = 2600
time_factor = np.where((weather['hour'] >= 8) & (weather['hour'] <= 20), 1.25, 0.95)
weekend_dip = np.where(weather['timestamp'].dt.dayofweek >= 5, 0.90, 1.0)

# AC/cooling load kicks in >35°C
ac_load = np.maximum(0, weather['temperature_2m'] - 35) * 18

# Industrial/commercial noise
noise = np.random.normal(0, 70, len(weather))

weather['demand_MW'] = base_load * time_factor * weekend_dip + ac_load + noise
weather['demand_MW'] = weather['demand_MW'].clip(1800, 4200)
print("✅ Demand column created. (Ready to swap with real data)")
```

```python
# 🧹 CELL 4: Clean, Convert to IST & Export
print("🧹 Cleaning & exporting...")

# Keep only needed columns
df = weather[['timestamp', 'demand_MW', 'temperature_2m', 'relative_humidity_2m', 'wind_speed_10m', 'cloud_cover']].copy()

# Convert UTC → IST
df['timestamp'] = df['timestamp'].dt.tz_localize('UTC').dt.tz_convert('Asia/Kolkata').dt.tz_localize(None)

# Sort & handle missing values (time-series safe)
df = df.sort_values('timestamp').reset_index(drop=True)
df = df.interpolate(method='linear').ffill().bfill()

# Export RapidMiner-ready CSV
OUTPUT_PATH = "ahmedabad_energy_weather_ready.csv"
df.to_csv(OUTPUT_PATH, index=False)
print(f"💾 Saved to: {OUTPUT_PATH}")
print(f"📊 Shape: {df.shape} | Date Range: {df['timestamp'].min()} → {df['timestamp'].max()}")
print(df.head())
```

```python
# 📊 CELL 5: Quick Validation Plot
print("📈 Plotting 7-day sample...")
sample = df[df['timestamp'].dt.month == 5].head(168)  # May week sample
plt.figure(figsize=(10,4))
plt.plot(sample['timestamp'], sample['demand_MW'], label='Demand (MW)', color='blue')
plt.plot(sample['timestamp'], sample['temperature_2m'], label='Temp (°C)', color='red', alpha=0.7)
plt.xticks(rotation=45)
plt.title("Ahmedabad: Demand vs Temperature (Sample Week)")
plt.legend()
plt.tight_layout()
plt.show()
```
