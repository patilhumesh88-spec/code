```python
import pandas as pd
```

```python
df=pd.read_csv('/content/NewspaperData.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df[['daily', 'sunday']].corr()
```

```python
import matplotlib.pyplot as plt
```

```python
daily_data=df['daily']
sunday_data=df['sunday']

plt.figure(figsize=(10,5))
plt.scatter(daily_data, sunday_data,color= 'red',alpha=0.5)
plt.title("DAILY VS SUNDAY")
plt.xlabel("DAILY")
plt.ylabel("SUNDAY")
plt.show()
```

```python
import statsmodels.formula.api as smf
model = smf.ols(formula='sunday ~ daily', data=df).fit()
```

```python
model.params
```

```python
newdata=pd.Series([200,300,500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
```

```python
data_pred
```

```python
model.predict(data_pred)
```

```python
model.rsquared
```
